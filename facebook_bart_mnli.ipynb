```python
from datasets import load_dataset

dataset = load_dataset("christophsonntag/OLID")
```

```python
import pandas as pd
df=dataset['train'].to_pandas()
```

```python
df_off=df.loc[df['subtask_a'].str.contains('OFF')]
df_noff=df.loc[df['subtask_a'].str.contains('NOT')]
```

```python
df.head()
```

```python
data = df[['tweet', 'subtask_a']]
data['subtask_a'].replace('OFF', 1, inplace=True)
data['subtask_a'].replace('NOT', 0, inplace=True)
```

## ZERO SHOT CLASSIFICATION using Hugging Face's built in pipeline

```python
data.head()
```

```python

import torch
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
labels = ["Severely Offensive", "Mildly Offensive", "Not Offensive"]
```

```python
# Example input text
text = "one day I will see the world"

result = classifier(text, candidate_labels=labels,multi_label=True)

# Print the results
print(result)
```

```python
predictions=[]
for text in data['tweet']:
  result=classifier(text,candidate_labels=labels,multi_label=True)

     # Extract the value for "Severely Offensive"
  severely_offensive_score = next(score for label, score in zip(result['labels'], result['scores']) if label == "Severely Offensive")
   # Add 1 to the prediction column if the score is greater than 0.8
  if severely_offensive_score > 0.8:
      predictions.append(1)
  else:
      predictions.append(0)

data['predictions']=predictions

data.head()
```

```python
 data.head()
```

```python
from sklearn.metrics import f1_score
```

```python
f1 = f1_score(data['subtask_a'], data['predictions'])
print(f"F1 Score: {f1}")
```
